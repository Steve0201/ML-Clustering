### Load data
Data contains some missing values, which is visible in the first few records. <br> Because there are no commas and multiple spaces between columns I had to load data in the following way.

1. In every line I substitute multiple spaces with single space
2. Then replace ' ' with ','
3. I drop last element, because it is empty (to be specific empty = '')


```python
with open('auto-mpg.data-original', 'r', newline='') as csv_file:
    reader = csv.reader(re.sub('\s+',' ',line).replace(' ',',') for line in csv_file)
    my_list = list(reader)
    my_list2 = []
    for m in my_list:
        my_list2.append(m[:-1])
```

### Create dataframe with data

```python
df = pd.DataFrame(my_list2)
cols = ['mpg','cylinders','displacement','horsepower','weight','acceleration','model year','origin','car name']
df.columns = cols
df.head(12)
# df
```

```python
##########################################################################
### Try to change all columns except the last one to the numeric records
### Check in what columns it is not possible
#########################################################################
for col in cols[:-1]:
    try:
        df[col] = pd.to_numeric(df[col])
    except:
        print('Problem to convert column:', col)
```

#### Error because of 'NA' string

```python
#################################################################
# Here I check what causes that I cannot make the column numeric
#################################################################

# df['mpg'] = pd.to_numeric(df['mpg'])
```

```python
###########################################
# Replace all NA with the mean mpg value
###########################################

mpg_row = df['mpg']
values = [float(i) for i in mpg_row if i != 'NA']
mpg_mean = np.round(np.mean(values),0)
print('Mean value of the mpg =', mpg_mean) # check if there are no np.nan        
```

```python
def change_data(x):
    if x == 'NA':
        return mpg_mean
    else:
        return x
    
df["mpg"]=df["mpg"].apply(change_data)
df.head(3)
```

```python
# Now there is no error
df['mpg'] = pd.to_numeric(df['mpg'])
```

```python
###########################################
# Again error because of 'NA' string
###########################################

# df['horsepower'] = pd.to_numeric(df['horsepower'])
```

```python
# count how many values in 'horsepower' column is 'NA'
count_na = 0
for i in df['horsepower']:
    if i == 'NA':
        count_na += 1
print('Number of values with NA =', count_na)
```

```python
###########################################
# Drop rows where are NA in 'horsepower'
###########################################

indexNames = df[df['horsepower'] == 'NA' ].index
df.drop(indexNames , inplace=True)
df.reset_index(drop=True,inplace=True)
```

```python
# Now there is no error
df['horsepower'] = pd.to_numeric(df['horsepower'])
```

```python
# Check if there are also some missing data
df.isnull().values.any()
```

# Clustering

```python
error_list = []
Ks = [i for i in range(1,11)]

for i in Ks:
    kmeans = KMeans(n_clusters = i).fit(df)
    error_list.append(kmeans.inertia_)
    
plt.figure(figsize=(8,6))   
plt.plot(Ks, error_list)
plt.plot(Ks, error_list,'bo')
plt.title('Elbow method')
plt.xlabel('clusters number')
plt.ylabel('error')
plt.show()
```

```python
y_predicted = KMeans(n_clusters = 3).fit_predict(df)
# y_predicted
```

```python
df['cluster'] = y_predicted
```

```python
df.head(3)
```

```python
# plt.style.use('dark_background')
plt.figure(figsize=(8,6))
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]

ind1 = 0
ind2 = 3

plt.scatter(df1[cols[ind1]],df1[cols[ind2]],color ='r')
plt.scatter(df2[cols[ind1]],df2[cols[ind2]],color ='b')
plt.scatter(df3[cols[ind1]],df3[cols[ind2]],color ='g')

plt.xlabel(cols[ind1])
plt.ylabel(cols[ind2])
plt.legend(['cluster1','cluster2','cluster3'])
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
# from sklearn.preprocessing import MinMaxScaler

df_scaled = pd.DataFrame([])
scaler = StandardScaler()


for col in cols[:-1]:
    scaled_features = scaler.fit_transform(df[[col]])
    df_scaled[col] = list(scaled_features.T)[0]
```

```python
df_scaled.tail()
```

```python
y_predicted = KMeans(n_clusters = 3).fit_predict(df_scaled)
```

```python
choosen_cols = ['mpg','displacement','horsepower','weight','acceleration']
num = 0
fig = plt.figure(figsize=(20,24))
y_predicted = KMeans(n_clusters = 3).fit_predict(df_scaled)
df_scaled['cluster'] = y_predicted
df11 = df_scaled[df_scaled.cluster==0]
df22 = df_scaled[df_scaled.cluster==1]
df33 = df_scaled[df_scaled.cluster==2]

for i in range(len(choosen_cols)):
    for j in range(len(choosen_cols)):
        if i != j and j>i:
            
            col1,col2 = choosen_cols[i],choosen_cols[j]

            num += 1
            ax = plt.subplot(5,2,int(num))
            
            plt.scatter(df11[col1],df11[col2],color ='r')
            plt.scatter(df22[col1],df22[col2],color ='b')
            plt.scatter(df33[col1],df33[col2],color ='g')

            plt.xlabel(col1)
            plt.ylabel(col2)
            plt.legend(['cluster1','cluster2','cluster3'])
            
plt.suptitle("Clustering for all data",fontsize=20)
plt.show()
```
