```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix
from scipy.stats import mode
```

```python
import pandas as pd
import numpy as np
```

## Preprocessing of data

```python
data = pd.read_csv("Cars.data",header=None)
```

```python
print('Size of our data before preprocessing: ',data.shape)
data.head()
```

First of all I want to add names of columns to our data.

```python
Columns = ['symboling','normalized_losses','make','fuel_type','aspiration','num_of_doors','body_style','drive_wheels','engine_location',
           'wheel_base','length','width','height','curb_weight','engine_type','num_of_cylinders','engine_size','fuel_system','bore',
           'stroke','compression_ratio','horsepower','peak_rpm','city_mpg','highway_mpg','price']

data.columns=Columns
data.head()
```

In our data missing value are marked with sign '?'.
Let's look how many miising value we have.

```python
for name in Columns:
    table = data[name]
    suma = 0
    for element in table:
        if element == '?':
            suma = suma + 1
    if suma > 0:
        print('Number of missing values in column ' + str(name) + ' :',suma)
```

As we can see there is a lot of missing values in column normalized_losses and small value of missing values in some different columns.
If I remove all rows with missing values I just remove 20% of our data. It is too much so I decided to replace this missing values in normalized_losses column by the integer mean value (because values in this column are integer)  and remove rows with different missing values.


```python
table = data['normalized_losses']
table2 = []

for element in table:
    if element != '?':
        table2.append(int(element))
mean = int(np.mean(table2))
new_values = []

for element in table:
    if element != '?':
        new_values.append(int(element))
    else:
        new_values.append(mean)

data['normalized_losses'] = new_values
```

Look again on missing values:

```python
for name in Columns:
    table = data[name]
    suma = 0
    for element in table:
        if element == '?':
            suma = suma + 1
    if suma > 0:
        print('Number of missing values in column ' + str(name) + ' :',suma)
        
```

```python
indexNames = data[(data['num_of_doors'] == '?')|(data['bore'] == '?')|(data['stroke'] == '?')|(data['horsepower'] == '?')|(data['peak_rpm'] == '?')|(data['price'] == '?')].index

data.drop(indexNames , inplace=True)
```

Look again on missing values:

```python
for name in Columns:
    table = data[name]
    suma = 0
    for element in table:
        if element == '?':
            suma = suma + 1
    if suma > 0:
        print('Number of missing values in column ' + str(name) + ' :',suma)
        
print('Size of our data after preprocessing: ',data.shape)
```

We do not have any results from cell above so we do not have missing values.

Now we can check types of our columns. 

```python
data.dtypes
```

In this columns where was missing values, except num_of_doors column, we have types object instead of numeric type.

```python
data["bore"] = pd.to_numeric(data["bore"])
data["stroke"] = pd.to_numeric(data["stroke"])
data["horsepower"] = pd.to_numeric(data["horsepower"])
data["peak_rpm"] = pd.to_numeric(data["peak_rpm"])
data["price"] = pd.to_numeric(data["price"])
```

```python
data.dtypes
```

```python
data.reset_index(drop=True,inplace=True)
data
```

## Find optimal k number of clusters

```python
data = load_iris()
Data = data.data
Result = data.target
Names = data.target_names
```

Thereâ€™s a method called the Elbow method, which is designed to help find the optimal number of clusters in a dataset.

We know that in our datasets there are only three results of flower type. So if the result of Elbow method will be $k=3$
that means in our data we have optimal value of results.

```python
Error =[]
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i).fit(Data)
    Error.append(kmeans.inertia_)
plt.plot(range(1, 11), Error)
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Error')
plt.show()
```

As we can see on the picture shape of the curve formed at $k=3$. That means in our data three groups of results are optimal.

## Clusters with k = 3

```python
Clusters = KMeans(n_clusters = 3).fit_predict(Data)

Labels = np.zeros_like(Clusters)

## Comparing number of cluster with actual number of cluster

for i in range(3):
    mask = (Clusters == i)
    Labels[mask] = mode(data.target[mask])[0]
```

```python
### Present the results
table = confusion_matrix(Result,Labels)
sns.heatmap(table.T, square=True, annot=True,cmap='RdYlGn',xticklabels=Names,yticklabels=Names)
plt.xlabel('true label')
plt.ylabel('predicted label')
plt.title("Comparison actual value with prediction value")
plt.show()

Accuracy = 0
for i in range(len(Result)):
    if Result[i] == Labels[i]:
        Accuracy = Accuracy + 1
        
Accuracy = Accuracy/len(Result)*100
print('Accuracy of clustering by k-means method: ',round(Accuracy,2),'%')
```

## Clusters k = 3 using each pair of columns

```python

Features = ['sepal length','sepal width','petal length','petal width']
plt.figure(figsize=(15,15))
num = 1
### SCATTER PLOT ###
for i in range(3):
    for j in range(i+1,4):
        
        NewData = Data[:,[i,j]]
        Clusters = KMeans(n_clusters = 3).fit_predict(NewData)
        
        Labels = np.zeros_like(Clusters)
        for k in range(3):
            mask = (Clusters == k)
            Labels[mask] = mode(data.target[mask])[0]    
            
        Accuracy = 0
        for l in range(len(Result)):
            if Result[l] == Labels[l]:
                Accuracy = Accuracy + 1
        Accuracy = Accuracy/len(Result)*100
        
        sub = '32' + str(num)
        ax = plt.subplot(int(sub))
        for  k in range(NewData.shape[0]):
            if Result[k] == 0:
                c = 'red'
                if Labels[k] == 0:
                    m = 'o'
                elif Labels[k] == 1:
                    m = 'v'
                elif Labels[k] == 2:
                    m = 's'
            elif Result[k] == 1:
                c = 'blue'
                if Labels[k] == 0:
                    m = 'o'
                elif Labels[k] == 1:
                    m = 'v'
                elif Labels[k] == 2:
                    m = 's'
            elif Result[k] == 2:
                c = 'green'
                if Labels[k] == 0:
                    m = 'o'
                elif Labels[k] == 1:
                    m = 'v'
                elif Labels[k] == 2:
                    m = 's'
            plt.scatter(NewData[k][0],NewData[k][1],color = c,marker = m)
        plt.xlabel('true label')
        plt.ylabel('predicted label')
        plt.title(Features[i] + ' and ' + Features[j] + ' accuracy: ' + str(round(Accuracy,1)) + '%') 
        num = num + 1

plt.show()

#### HEAT MAP ####
plt.figure(figsize=(15,15))
num = 1

for i in range(3):
    for j in range(i+1,4):
        
        NewData = Data[:,[i,j]]
        Clusters = KMeans(n_clusters = 3).fit_predict(NewData)

        Labels = np.zeros_like(Clusters)
        for k in range(3):
            mask = (Clusters == k)
            Labels[mask] = mode(data.target[mask])[0]
        
        Accuracy = 0
        for l in range(len(Result)):
            if Result[l] == Labels[l]:
                Accuracy = Accuracy + 1
        Accuracy = Accuracy/len(Result)*100
        
        sub = '32' + str(num)
        ax = plt.subplot(int(sub))
        
        table = confusion_matrix(Result,Labels)
        sns.heatmap(table.T, square=True, annot=True,cmap='RdYlGn',xticklabels=Names,yticklabels=Names)
        plt.xlabel('true label')
        plt.ylabel('predicted label')
        plt.title(Features[i] + ' and ' + Features[j] + ' accuracy: ' + str(round(Accuracy,1)) + '%')
        num = num + 1
plt.show()
```
